```python
katok = ['','','','','','']

def insert_data(position, friend):
    
    katok.append(None)
    kLen = len(katok)
    
    for i in range(kLen-1, position, -1):
        katok[i] = katok[i-1]
        katok[i-1] = None
    katok[position] = friend
```

```python
insert_data(2,'')
```
